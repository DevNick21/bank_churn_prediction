```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
raw_data = pd.read_csv("Churn_Modelling.csv")
raw_data
```

```python
df = raw_data.drop(columns=['RowNumber', 'CustomerId', 'Surname'])
df
```

```python
df.describe()
```

```python
df.head()
```

```python
dummies_location = pd.get_dummies(df['Geography'], dtype= int)
dummies_location
```

```python
dummies_gender = pd.get_dummies(df['Gender'], dtype=int)
dummies_gender
```

```python
df = pd.concat([df, dummies_location, dummies_gender], axis=1)
df
```

```python
df = df.drop(columns=['Geography', 'Gender'])
df
```

```python
df.columns
```

```python
column_values = ['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
                 'IsActiveMember', 'EstimatedSalary', 'Exited', 'France', 'Germany',
                 'Spain', 'Female', 'Male']
```

```python
df.columns = column_values
df
```

```python
df = df.reindex(columns=['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
                    'IsActiveMember', 'EstimatedSalary', 'France', 'Germany',
                    'Spain', 'Female', 'Male','Exited'])
```

```python
df
```

```python
x = df.drop(columns='Exited')
x
```

```python
y = df['Exited']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train_scaled = scaler.transform(x_train)
x_val_scaled = scaler.transform(x_val)
x_test_scaled = scaler.transform(x_test)
```

```python
x_test_scaled.shape
```

```python
y_train.shape
```

```python
x_val_scaled.shape
```

```python
x_test_scaled.shape
```

```python
np.savez('Bank_Churn_train', inputs=x_train_scaled, targets=y_train)
np.savez('Bank_Churn_validation', inputs=x_val_scaled, targets=y_val)
np.savez('Bank_Churn_test', inputs=x_test_scaled, targets=y_test)
```

```python
import pickle

with open('scaler', 'wb') as f:
    pickle.dump(scaler, f)
```
